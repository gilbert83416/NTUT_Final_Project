```python
import pandas as pd
# 刪除第0列(英文，非資料)
Q1_109 = pd.read_csv("109年度第一季實價登錄.csv", encoding = "utf-8").drop([0])
Q2_109 = pd.read_csv("109年度第二季實價登錄.csv", encoding = "utf-8").drop([0])
Q3_109 = pd.read_csv("109年度第三季實價登錄.csv", encoding = "utf-8").drop([0])
Q4_109 = pd.read_csv("109年度第四季實價登錄.csv", encoding = "utf-8").drop([0])
# Q1_109
```

```python
datas = pd.concat([Q1_109, Q2_109, Q3_109, Q4_109], axis = 0, ignore_index = True)
datas
```

```python
# 查詢各欄位空值
s = datas.isnull().sum()
s[s>0].sort_values(ascending = False)
```

```python
amend_datas = datas
# 1. columnB: 只有車位和土地的數量(要drop)
print(amend_datas["交易標的"].isin(["車位"]))
# print(amend_datas["交易標的"].isin(["土地"]).sum())
```

```python
# 交易筆棟數-資料分割
amend_datas["交易筆棟數-土地"] = datas["交易筆棟數"].str.split("土地|建物", expand=True)[1]
amend_datas["交易筆棟數-建物"] = datas["交易筆棟數"].str.split("建物|車位", expand=True)[1]
amend_datas["交易筆棟數-車位"] = datas["交易筆棟數"].str.split("車位", expand=True)[1]
amend_datas = amend_datas.drop(["交易筆棟數"], axis=1)
# amend_datas
```

```python
# 2. columnE: 刪除用不到的column(F:非都市土地使用分區、G:非都市土地使用編定)
amend_datas = amend_datas.drop(["非都市土地使用分區","非都市土地使用編定"], axis=1)
amend_datas
```

```python
# amend_datas.to_csv("0422-1.csv", encoding="utf-8", index=False)
amend_datas.iloc[19569]
```

```python
# 刪除<建築完成年月>有異data : iloc[19569]= 076 03
amend_datas = amend_datas.drop(labels=19569,axis = 0)
amend_datas.iloc[19569]
```

```python
amend_datas.iloc[19569]
```

```python
# 3. columnＬ（建物型態）分類：住宅大樓、公寓、套房、華廈、透天、商辦
house_types = ['住宅大樓(11層含以上有電梯)','公寓(5樓含以下無電梯)','套房(1房1廳1衛)','華廈(10層含以下有電梯)','透天厝']
amend_datas[amend_datas['建物型態'].isin(house_types)]
amend_datas = amend_datas[amend_datas['建物型態'].isin(house_types)]
amend_datas
```

```python
transaction = ["建物","房地(土地+建物)","房地(土地+建物)+車位"]
amend_datas = amend_datas[amend_datas["交易標的"].isin(transaction)]
amend_datas
```

```python
materials = ['土木造','土造','土磚石混合造','木造', '石造', '磚造', '加強磚造', '壁式預鑄鋼筋混凝土造', '鋼筋混凝土造', '鋼骨混凝土造']
amend_datas[amend_datas["主要建材"].isin(materials)]
amend_datas = amend_datas[amend_datas["主要建材"].isin(materials)]
amend_datas
```

```python
# 4. columnＮ（主要建材）分類：1[土木造、土造、土磚石混合造、木造、石造]、2[磚造]、3[加強磚造]、4[壁式預鑄鋼筋混凝土造、鋼筋混凝土造]、5[鋼骨鋼筋混凝土造]、6[鋼骨混凝土造]
material_type = {'土木造':1, '土造':1,'土磚石混合造':1,'木造':1, '石造':1, '磚造':2, '加強磚造':3, '壁式預鑄鋼筋混凝土造':4, '鋼筋混凝土造':4, '鋼骨混凝土造':4}
def material(m):
  return material_type[m]
amend_datas['主要建材'] = amend_datas['主要建材'].apply(material)
amend_datas
```

```python
# 5. columnＨ（交易年月）：取年 ＃保留非109年的成交建案
amend_datas['交易年月日']
trade_year =[y[:3] for y in amend_datas['交易年月日']]
trade_year
amend_datas['交易年份'] = trade_year
amend_datas
```

```python
amend_datas.reset_index(drop=True, inplace=True)
amend_datas
```

```python
# 6. columnＯ（建築完成年月）轉成屋齡（小數）/ 空值補？ 參考鐵達尼
amend_datas["建築完成年月"] = amend_datas["建築完成年月"].fillna('0751029')
for i in range(0, len(amend_datas["建築完成年月"])):
  amend_datas["建築完成年月"][i] = str(int(amend_datas["建築完成年月"][i])+ 19110000) 
    # print(i, amend_datas["建築完成年月"][i])
# amend_datas.to_csv('0422-2.csv', encoding = 'utf-8')
amend_datas
```

```python
for i in range(0, len(amend_datas["建築完成年月"])):
  # amend_datas["建築完成年月"][i] = (str(amend_datas["建築完成年月"][i])[0:4]+','+str(amend_datas["建築完成年月"][i])[4:6] + ','+str(amend_datas["建築完成年月"][i])[6:8])
  amend_datas["建築完成年月"][i] = (str(2021-int(amend_datas["建築完成年月"][i][0:4])))
amend_datas
```

```python
# 7. columnＴ（建築現況格局-隔間）：整列刪除
amend_datas = amend_datas.drop(["建物現況格局-隔間"], axis=1)
```

```python
# 樓層轉數字
levels = { "一層":1,"二層":2,"三層":3,"四層":4,"五層":5,"六層":6,"七層":7,"八層":8,"九層":9,"十層":10,
    "十一層":11,"十二層":12,"十三層":13,"十四層":14,"十五層":15,"十六層":16,"十七層":17,"十八層":18,"十九層":19,"二十層":20,
    "二十一層":21,"二十二層":22,"二十三層":23,"二十四層":24,"二十五層":25,"二十六層":26,"二十七層":27,"二十八層":28,"二十九層":29,"三十層":30,
    "三十一層":31,"三十二層":32,"三十三層":33,"三十四層":34,"三十五層":35,"三十六層":36,"三十七層":37,"三十八層":38,"三十九層":39,"四十層":40,
    "四十一層":41,"四十二層":42,}
amend_datas["總樓層數"] = amend_datas["總樓層數"].replace(levels)
# amend_datas
```

```python
# 8. columnＵ（有無管理組織）：轉布林值：無（0）、 有（1）
management  = {"無":0, "有":1}
amend_datas["有無管理組織"] = amend_datas["有無管理組織"].replace(management)
```

```python
# 9. columnＷ（單價元平方公尺）：換成坪（1m^2 = 0.3025 坪）
amend_datas["單價元/坪"] = amend_datas["單價元平方公尺"].astype(float) /0.3025
amend_datas
```

```python
# on-hot encoding
amend_datas = amend_datas.drop(["交易標的"], axis=1)
amend_datas = amend_datas.drop(["都市土地使用分區"], axis=1)
amend_datas = amend_datas.drop(["交易年月日"], axis=1)
amend_datas = amend_datas.drop(["主要用途"], axis=1)
amend_datas = amend_datas.drop(["車位類別"], axis=1)
amend_datas = amend_datas.drop(["車位移轉總面積平方公尺"], axis=1)
amend_datas = amend_datas.drop(["車位總價元"], axis=1)
amend_datas = amend_datas.drop(["備註"], axis=1)
amend_datas = amend_datas.drop(["移轉層次"], axis=1)
amend_datas = amend_datas.drop(["車位移轉總面積(平方公尺)"], axis=1)
amend_datas = amend_datas.drop(["主建物面積"], axis=1)
amend_datas = amend_datas.drop(["附屬建物面積"], axis=1)
amend_datas = amend_datas.drop(["陽台面積"], axis=1)
amend_datas = amend_datas.drop(["電梯"], axis=1)
amend_datas = amend_datas.drop(["移轉編號"], axis=1)


amend_datas = pd.get_dummies(amend_datas, columns=['鄉鎮市區'])
amend_datas = pd.get_dummies(amend_datas, columns=['建物型態'])
amend_datas.to_csv('0424.csv', encoding= 'utf-8')
```
